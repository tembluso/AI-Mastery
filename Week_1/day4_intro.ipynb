Week 1 – Day 4: Data Visualization

Objective of the day
Learn how to visualize data with matplotlib and seaborn. By the end of today, you’ll create plots to make patterns (like survival differences) immediately clear.

1. Setup

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load Titanic again
import pandas as pd
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
df = pd.read_csv(url)

# Basic histogram (age distribution)

plt.hist(df["age"].dropna(), bins=20)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age distribution")
plt.show()

# Count plot (survival counts)

sns.countplot(x="survived", data=df)
plt.title("Survival counts")
plt.show()

# Bar Plot (survival rate by sex)

sns.barplot(x="sex", y="survived", data=df)
plt.title("Survival rate by sex")
plt.show()

# Facet plots (two variables at once)

sns.barplot(x="pclass", y="survived", hue="sex", data=df)
plt.title("Survival by class and sex")
plt.show()

```
